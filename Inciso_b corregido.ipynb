# Gráficas del problema de Dirichlet.

Para resolver la ecuación de LaPlace:
$$\frac{\partial u}{\partial t} = \frac{\partial^2 u}{\partial x^2},\;\;\;\
0< x < 1 \;\;\, \;\ 0 < y < 1 $$
Para las condiciones de frontera \\
b)  $$u(0,y)=10y, ~~ u_x(1,y)=-1, \;\;\; $$
$$u(x,0)=u(x,1)=0\;\;\; $$

Como ya sabemos la serie de Fourier resultante la colocamos:
$$\sum_{n=1}^{\infty} [\frac{-20 (-1)^n}{n \pi}(cosh(xn\pi))+\frac{2[(-1)^n-1]+20(-1)^n(n \pi)\sinh (n\pi)}{n^2 \pi^2 \cosh (n \pi)}senh(yn\pi)]sen(xn\pi) .$$

```python
from pylab import *
from matplotlib import animation,rc
from IPython.display import HTML 
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
```

```python
x = linspace(0,1.0,1000)
y = linspace(0,1.0,1000)
```

Usamos de manera independiente nuestros **An** y **Bn**: \\
**An**= $$\frac{-20 (-1)^n}{n \pi}$$ \\
**Bn**= $$\frac{2[(-1)^n-1]+20(-1)^n(n \pi)\sinh (n\pi)}{n^2 \pi^2\cosh (n \pi)}$$ \\

```python
def u_x_y(x,y):
    N = 8
    f = zeros ((len(x),len(y)))
    for i in range(len(x)):
        for n in range(1,N):
          a = (-20*(-1)**n)/(n*pi)
          b = (2*((-1)**n-1)+20*(n*pi)*sinh(n*pi)*(-1)**n)/(n**2*pi**2*cosh(n*pi))
          f[:,i] += sin(n*pi*y)*(a*cosh(n*pi*x[i])+b*sinh(n*pi*x[i]))
    return f
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(15,12))
ax = fig.add_subplot(111, projection="3d")

X,Y = np.meshgrid(x,y)

U = u_x_y(x,y)

plt.xlabel('$x$'); plt.ylabel('$y$'); 
plt.title("u(x,y)")
ax.plot_surface(X, Y, U, cmap="inferno");
```
