# Gráficas del problema de Dirichlet.


Para resolver la ecuación de LaPlace:
$$\frac{\partial u}{\partial t} = \frac{\partial^2 u}{\partial x^2},\;\;\;\
 0 < x < 1 \;\;\, \;\ 0< y < 1$$
Para las condiciones de frontera \\
a) $$ u(0,y)=u(1,y)=0; u(x,0)=100, u(x,1)=200 $$

Como ya sabemos la serie de Fourier resultante la colocamos:
$$\sum_{n=1}^{\infty} \frac{200(1-(-1))^n}{n\pi}[(cosh(yn\pi))+\frac{2-cosh(n\pi)}{senh(n\pi)}senh(yn\pi)]sen(xn\pi) .$$

```python
from pylab import *
from matplotlib import animation,rc
from IPython.display import HTML 
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
```

```python
x= linspace(0,1,1000)
y= linspace(0,1,1000)
```

Usamos de manera independiente nuestros **An** y **Bn**: \\
**An**= $$\frac{200(1-(-1))^n}{n\pi}$$ \\
**Bn**= $$\frac{200(1-(-1)^n}{(2-cosh(n\pi))}$$ \\

```python
def u_x_y(x,y):
    N = 10
    f = zeros ((len(x),len(y)))
    for j in range(len(y)):
        for n in range(1,N):
          a = (200*(1+(-1)**(n+1)))/(n*pi)
          b = (200*(1+(-1)**(n+1))*(2-cosh(n*pi)))/((n*pi)*sinh(n*pi))
          f[:,j] += sin(n*pi*x)*(a*cosh(n*pi*y[j])+b*sinh(n*pi*y[j]))
    return f
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(16,10))
ax = fig.add_subplot(111, projection="3d")

X,Y = np.meshgrid(x,y)

U = u_x_y(x,y)

plt.xlabel('$x$'); plt.ylabel('$y$'); 
plt.title("u(x,y)")
ax.plot_surface(X, Y, U, cmap="inferno"); 
ax.plot_surface(X, Y, U, cmap="inferno"); 
```
